    Types of recommendation system

==> Popularity Based
    It keeps a track of view counts for each movie/video and then lists movies based on views in descending order.

==> Content Based
    This type of recommendation systems, takes in a movie that a user currently likes as input. Then it analyzes the contents of the movie to find out      other movies which have similar content. Then it ranks similar movies according to their similarity scores and recommends the most relevant movies      to the user.

==>Collaborative filtering
   In other words, the recommendations get filtered based on the collaboration between similar user’s preferences.

   In this project we are going to implement content based recommendation system.

Data Cleaning

Here we want to collect only the needed tags wich is significant in recommendations.


```python
import ast
```

```python
def convert(obj):
    List=[]
    for i in ast.literal_eval(obj):
        List.append(i['name'])
    return List   
```

```python
movies_df['genres'] = movies_df['genres'].apply(convert)
```

```python
movies_df['keywords'] = movies_df['keywords'].apply(convert)
```

```python
def convert_cast(obj):
    List=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter != 3:
            List.append(i['name'])
            counter += 1 
        else:
            break
    return  List       
```

```python
movies_df['cast'] = movies_df['cast'].apply(convert_cast)
```

```python
def convert_director(obj):
    List=[]
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            List.append(i['name'])
    return List        
```

```python
movies_df['crew'] = movies_df['crew'].apply(convert_director)
```

```python
movies_df.head()
```

```python
movies_df['overview'] = movies_df['overview'].apply(lambda x:x.split())
```

```python
movies_df.head()
```

```python
movies_df['genres'] = movies_df['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies_df['keywords'] = movies_df['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies_df['cast'] = movies_df['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies_df['crew'] = movies_df['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies_df.head()
```

```python
movies_df['tags'] = movies_df['overview']+movies_df['genres']+movies_df['cast']+movies_df['crew']+movies_df['keywords']
```

```python
movies_df.head()
```

```python
final_df = movies_df[['id','title','tags']]
final_df.head()
```

```python
final_df['tags'] = final_df['tags'].apply(lambda x:' '.join(x))
```

```python
final_df['tags'][0]
```

```python
final_df['tags'] = final_df['tags'].apply(lambda x:x.lower())
```

```python
final_df
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

 Porter stemming algorithm, is a popular algorithm used in natural language processing and information retrieval to reduce words to their root or base form, known as the "stem." The primary goal of stemming is to transform words with similar meanings into a common base form, which helps in improving text analysis and information retrieval tasks.

```python
def stem(text):
    List = []
    for i in text.split():
        List.append(ps.stem(i))
    return " ".join(List)
```

```python
final_df['tags'] = final_df['tags'].apply(stem)
```

```python
final_df['tags']
```

Content Based Filtering

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(stop_words='english')
vectors = cv.fit_transform(final_df['tags']).toarray()
vectors
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_similarity(vectors).shape
```

```python
similarity = cosine_similarity(vectors)
similarity
```

```python
sorted(list(enumerate(similarity[0])),reverse = True,key = lambda x:x[1])[1:20]
```

```python
def recommend(movie):
    movie_index = movies_df[movies_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)),reverse = True,key = lambda x:x[1])[1:20]
    
    for i in movie_list:
        print(final_df.iloc[i[0]].title)
```

```python
recommend('Street Kings')
```

```python
recommend('Titanic')
```
